```python
input_sequences=[]

for sentence in faqs.split('\n'):
  tokenized_sentence = tok.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

This will become out I/P data and from this we will make O/P dataset, as we can see the size is different for every list and we cannot give this as input , we have to make the size equal so we will add zero padding to it.


```python
input_sequences
```

we will find the max len of the list biggest len and add that many zero to make everything on the same length

```python
max_len = max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences,maxlen= max_len,padding='pre')
```

```python
X = padded_input_sequences[:,:-1]
#ALL the rows and columns except the last
```

```python
X
```

```python
y = padded_input_sequences[:,-1]
y
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=283)
```

Ist layer = Embedding Layer = Converts each word to dense vector

2nd layer = lstm

Dense layer

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(283,100,input_length=56))
model.add(LSTM(150))
model.add(Dense(283,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```
